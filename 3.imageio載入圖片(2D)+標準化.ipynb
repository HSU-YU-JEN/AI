照片數量 or 添加channel

```python
batch_size = 4 # 三張照片
batch = torch.zeros(batch_size, 3, 256, 256, dtype=torch.uint8) # (3照片，RGB，高H，寬W，8-bit-integer)
```

(主要)載入圖片並標號儲存

```python
import torch
import torch.nn as nn
import os # 工作目錄
import imageio # 處理numpy圖片的

data_dir = os.getcwd() # 目前的工作目錄 getcwd() current working directory

filenames = [name for name in os.listdir(data_dir) if os.path.splitext(name)[-1] == '.png'] # 回傳data_dir內的資料 若name(2.)切割出來倒數第一個為'.png' 則返還成name儲存
# os.listdir()可以取得指定目錄中所有的檔案與子目錄名稱
# os.path.splitext()為切割path中文字
     
for i, filename in enumerate(filenames): # i為編號 filename為上面檔名有.png部分的檔案
  img_arr = imageio.imread(os.path.join(data_dir, filename)) # 抓(路徑+檔名.png) 用array存
  img_t = torch.from_numpy(img_arr) # 轉tensor
  img_t = img_t.permute(2, 0, 1) # 變(channel 高H 寬W)
  img_t = img_t[:3] # 取前面3個(channel 高H 寬W) 因為後面有一個alpha channel
  batch[i] = img_t # (batch_size channel 高H 寬W) batch[1]為第一張圖片的tensor 
# enumerate(跌代物, 編號) 前者輸入一個可迭代的對象、比如說 List；後者輸入開始的起點編號，為數字，若不設定時從 0 開始 
# os.path.join連接path文字
```

標準化(nn訓練效果會比較好)


法1. 直接壓縮成255

```python
# batch (batch_size channel 高H 寬W)
batch = batch.float() # 全轉float
batch /= 255.0 # 全除255
```

法2. 把數值常態分布化

```python
n_channels = batch.shape[1] # channel數量
for c in range(n_channels): # for 取 c 從 0 到 2
  mean = torch.mean(batch[:, c]) # 取平均值 :為全部的data
  std = torch.std(batch[:, c]) # 取標準差
  batch[:, c] = (batch[:, c] - mean) / std # 標準化公式
```
